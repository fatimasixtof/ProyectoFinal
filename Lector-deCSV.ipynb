```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv('resultados_hiperconsumismo.csv', skiprows=6)

df = df.rename(columns={
    "[step]": "Tick",
    "count turtles with [color-tendencia = tendencia-global]": "Usuarios_Tendencia",
    "tasa-de-aburrimiento": "Tasa_Aburrimiento",
    "influencia-externa": "Influencia_Externa"
})

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Tick", y="Usuarios_Tendencia", hue="Tasa_Aburrimiento", palette="flare", linewidth=2)
plt.title('Adopción de la Tendencia Global según la Tasa de Aburrimiento')
plt.xlabel('Tiempo (Ticks)')
plt.ylabel('Cantidad de Usuarios')
plt.legend(title="Tasa de Aburrimiento")
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Tick", y="Usuarios_Tendencia", hue="Influencia_Externa", palette="viridis", linewidth=2)
plt.title('Impacto de la Industria en la Tendencia')
plt.xlabel('Tiempo (Ticks)')
plt.ylabel('Cantidad de Usuarios')
plt.legend(title="Influencia Externa (%)")
plt.show()
```
